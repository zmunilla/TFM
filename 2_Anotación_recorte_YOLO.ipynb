IMPORTACIÓN LIBRERÍAS REQUERIDAS

```python
import os
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

APLICACIÓN DEL MODELO DE DETECCIÓN Y RECORTE A UNA CARPETA CON NUEVAS FOTOS

```python
# Cargo el mejor modelo que ya tengo entrenado
model = YOLO('/content/drive/MyDrive/TFM/modelos/best8.pt')

# Defino el valor de la confianza y de la iou o solapamiento

model.conf = 0.45
model.iou = 0.5

# Ruta a carpetas de entrada/salida

input_folder = '/content/drive/MyDrive/TFM/Lado/lado1'
output_folder = '/content/drive/MyDrive/TFM/Lado/lado_crop'

for file_name in os.listdir(input_folder): # Bucle que recorre las imágenes
    if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        input_path = os.path.join(input_folder, file_name) # Path de cada imagen

        # Transformación imágen
        image = cv2.imread(input_path) # Convierto imagen a matriz numpy

        # Predicción
        results = model(input_path) # anotación de bounding boxes en imagen

        for i, result in enumerate(results): # Bucle de las detecciones
            boxes = result.boxes.xyxy  # Coordenadas del bounding box
            confidences = result.boxes.conf  # Confianza de cada predicción

            # Recorte
            for j, (box, conf) in enumerate(zip(boxes, confidences)): # Bucle de las
            #bounding boxes detectadas en una imagen
                x1, y1, x2, y2 = map(int, box) # Conversión coordenadas a enteros

                # Recorte de imagen usando las coordenadas con el comando image de CV2
                cropped = image[y1:y2, x1:x2]

                # Ruta para guardar la imagen recortada
                cropped_output_path = os.path.join(output_folder,
                                                   f"{file_name.split('.')[0]}_crop_{j+1}.jpg")

                # Almacenamiento imagen recortada
                cv2.imwrite(cropped_output_path, cropped)

print("Finalizado ✂️")

```
